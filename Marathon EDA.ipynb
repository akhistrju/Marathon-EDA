```python
#combining the filters
```

```python
df2=df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020)&(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.info()
```

```python
#remove USA from event name
```

```python
df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#cleaning age
```

```python
df2['athelete_age'] = 2020-df2['Athlete year of birth']
```

```python
#remove h from athlete performance
```

```python
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#drop columns Athlete year of birth, Athlete age category
```

```python
df2 = df2.drop(['Athlete year of birth', 'Athlete age category','Athlete club','Athlete country'], axis =1)
```

```python
df2.head(5)
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check for duplicates
```

```python
df2[df2.duplicated()==True]
```

```python
#reset index
```

```python
df2 = df2.reset_index(drop=True)
```

```python
#fix types
```

```python
df2.dtypes
```

```python
df2['athelete_age']=df2['athelete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#rename columns
```

```python
# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed        float64
# Athlete ID                     int64
# athelete_age                   int64
```

```python
df2 = df2.rename(columns ={'Year of event':'year',
                          'Event dates':'race_day',
                          'Event name':'race_name',
                          'Event distance/length':'race_length',
                          'Event number of finishers':'race_number_of_finishers',
                          'Athlete performance':'athlete_performance',
                          'Athlete gender':'athlete_gender',
                          'Athlete average speed':'athlete_average_speed',
                          'Athlete ID':'athlete_id'})
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athelete_age','athlete_performance','athlete_average_speed']]
```

```python
df3.head()
```

```python
#what age groups are the best in the 50mil Race(20+races min)
```

```python
df3.query('race_length == "50mi"').groupby('athelete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
#what age groups are the worst in the 50mil Race(10+races min)
```

```python
df3.query('race_length == "50mi"').groupby('athelete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>9').head(15)
```

```python
#seasons, is average speed slower in summer than winter?
#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split between 2 decimals
```

```python
df3['race_month']= df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(5)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x:'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df3.tail(5)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
